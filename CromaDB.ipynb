Create ChromaDB client Object

```python
import chromadb

client = chromadb.Client()
```

Create ChromaDB collection

```python
collection = client.create_collection(name="my_collection")
```

Add data to ChromaDB collection

```python
Artificial_Inteligent = """
Artificial Intelligence (AI) enables machines to mimic human intelligence, learning from data to make decisions, solve problems, and automate tasks.
It powers technologies like chatbots, recommendation systems, and self-driving cars, transforming industries and everyday life.
AI combines algorithms, data, and computing power to deliver smart, adaptive, and efficient solutions.
"""

India_Cricket = """
India’s cricket team is one of the most successful and passionate in the world,
known for legendary players like Sachin Tendulkar and Virat Kohli. Representing billions of fans,
India has won multiple ICC tournaments, including the 1983 and 2011 World Cups, and continues to dominate across all formats globally.
"""

Cristiano_Ronaldo = """
Cristiano Ronaldo is a legendary Portuguese footballer known for his incredible skill, speed, and goal-scoring ability.
He has won numerous titles, including five Ballon d'Or awards and multiple UEFA Champions Leagues.
Playing for clubs like Manchester United, Real Madrid, and Al-Nassr, Ronaldo remains an icon in global football history.
"""

SriLnka_Politics = """
In 2025, Sri Lanka's political landscape underwent significant transformation.
Anura Kumara Dissanayake, leading the National People's Power (NPP) alliance, was elected president in September 2024, marking a shift towards leftist governance.
The NPP secured a supermajority in the November 2024 parliamentary elections, enabling substantial policy reforms. In May 2025, the NPP continued its momentum by winning the majority in local government elections, further consolidating its political influence.
"""
```

```python
collection.add(
    documents = [Artificial_Inteligent, India_Cricket, Cristiano_Ronaldo, SriLnka_Politics],
    metadatas = [{"source": "AI"},{"source": "cricket"},{'source':'ronaldo'},{"source":"politics"}],
    ids = ["id1", "id2", "id3", "id4"]
)
```

Similarity search

```python
results = collection.query(
    query_texts=["Technology transforms lives, connects worlds, and drives global innovation forward."],
    n_results=1
)

results
```

```python
collection.count()
```

#CRUD operations on Vector Database

Add data

```python
Construction = """
Construction is the process of building infrastructure like homes, roads, and skyscrapers.
It involves planning, designing, and assembling materials using skilled labor and machinery.
The industry drives economic growth, creates jobs, and shapes urban development.
Safety, sustainability, and technology are increasingly important in modern construction practices around the world.
"""

Tourism_Industry = """
The tourism industry is a major global economic driver, connecting people to diverse cultures, landmarks, and experiences.
It generates employment, boosts local economies, and promotes cultural exchange.
From eco-tourism to luxury travel, the sector adapts to trends and sustainability demands, making it vital for global development and cross-cultural understanding.
"""
```

```python
collection.add(
    documents = [Construction, Tourism_Industry],
    metadatas = [{"source": "construction"},{"source": "tourism"}],
    ids = ["id5", "id6"]
)
```

```python
collection.count()
```

```python
results = collection.query(
    query_texts=["Exploring new places boosts culture, economy, and global human connection."],
    n_results=1
)

results
```

Read data

```python
read = collection.get()
read
```

```python
read = collection.get(ids=["id1", "id5"])
read
```

Update data

```python
collection.update(
    ids=["id3"],
    documents=["Kumar Sangakkara, legendary Sri Lankan cricketer, was a brilliant batsman, sharp wicketkeeper, and respected leader in international cricket history."],
    metadatas=[{"source": "sanga"}],
)
```

```python
read = collection.get(ids=["id3"])
read
```
